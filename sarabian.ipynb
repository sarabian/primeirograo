```python
import os
import pandas as pd
import plotly.express as px
#depois de enterdermos o desafio proposto
#percorrer o sistema operacional em busca da pasta vendas procurar os arquivos
#importar base de dados de vendas
#tratar e compilar os dados
#calcular produto mais vendido
#calcular produto mais faturou
#calcular loja que mais faturou  dashbord
```

```python
#vamos utilizar a variavel pra armazenar o arquivo 
lista_arquivo=os.listdir("/content/drive/MyDrive/Vendas")
tabela_total=pd.DataFrame()
#estrutura de repeticao vamos utilizar o for
for arquivo in lista_arquivo:
    if "vendas" in arquivo.lower():
    
    #vamos importar os arquivos  
      tabela=pd.read_csv(f"/content/drive/MyDrive/Vendas/{arquivo}")    
      tabela_total=tabela_total.append(tabela)
display(tabela_total)
```

```python
#vamos compiar a tabela
tabela_produtos=tabela_total.groupby('Produto').sum()
tabela_produtos=tabela_produtos[['Quantidade Vendida']].sort_values(by='Quantidade Vendida',ascending=False)
display(tabela_produtos)
```

```python
tabela_total['faturamento']=tabela_total['Quantidade Vendida']*tabela_total['Preco Unitario']

tabela_faturamento=tabela_total.groupby('Produto').sum()
tabela_faturamento=tabela_faturamento[['faturamento']].sort_values(by='faturamento',ascending=False)
display(tabela_faturamento)
```

```python
tabela_loja=tabela_total.groupby('Loja').sum()
tabela_loja=tabela_loja[['faturamento']]
display(tabela_loja)
```

```python
grafico=px.bar(tabela_loja,x=tabela_loja.index,y='faturamento')
grafico.show()
```
